```python
# Group by speaker and new.vowel.class, then count occurrences of 'COLT' and 'CULT'
vowel_counts = df_subset.groupby(['speaker', 'new.vowel.class'])['new.vowel.class'].count().unstack(fill_value=0)

# Rename columns for clarity
vowel_counts = vowel_counts.rename(columns={'THOUGHT': 'COLT_count', 'LOT': 'CULT_count'})

# Count speakers with either COLT or CULT count as zero
zero_count_speakers = vowel_counts[(vowel_counts['COLT_count'] == 0) | (vowel_counts['CULT_count'] == 0)]

print(vowel_counts)
print("\nNumber of speakers with either COLT or CULT count as zero:", len(zero_count_speakers))
```
